```python
import cv2
import numpy as np
import os
from random import suffle
from tqdm import tqdm
import tensorflow as tf
import matplotlib.pyplot as plt
import tflearn
from tflearn.layers.conv import conv_2d,max_pool_2d
from tflearn.layers.core import input_data,dropout,fully_connected
from tflearn.layers.estimator import regression
#convolution(pass filter multiply /widow size take take no only) activation(remove -ve) polling(take max) arrane the no in
#row than pass it to fully connected layer,loss function
```

```python
TRAIN_DIR = 'train'
TEST_DIR = 'test'
IMG_SIZE = 50
LR = 1e-3
MODEL_NAME = 'dogs-vs-cats-convet'
def create_label(image_name):
    word_label = image_name.spli('.')[-3]
    if word_label =='cat':
        return np.array([1,0])
    elif word_label =='dog':
        return np.array([0,1])

def create_train_data():
    training_data =[]
    for img in tqdm(os.listdir(TRAIN_DIR)):
        path = os.path.join(TRAIN_DIR,img)
        img_data = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
        img_data = cv2.resize(img_data,(img_data),create_label(img)])
    suffle(training_data)
    np.save('train_data.npy',training_data)
    return training_data

def create_test_data():
    tsting_data=[]
    for img in tqdm(os.listdir(TEST_DIR)):
        path = os.path.join(TEST_DIR,img)
        img_num = img.split('.')[0]
        img_data = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
        img_data = cv2.resize(img_data,(IMG_SIZE,IMG_SIZE))
        testing_data.append([np.array(img_data),img_num])
    suffle(testing_data)
    np.save('test_data.npy',testing_data)
    return testing_data

train_data=create_train_data()
test_data=create_test_data()

train=train_data[:-500]
test=test_data[-500:]

X_train = np.array(i[0] for i in train]).reshape(-1,IMG_SIZE,IMG_SIZE,1)
y_train = [i[1] for i in train]

X_test = np.array(i[0] for i in test]).reshape(-1,IMG_SIZE,IMG_SIZE,1)
y_test = [i[1] for i in test]

#model building
tf.reset_default_graph()
convert = input_data(shape=[None,IMG_SIZE,IMG_SIZE,1],name='input')
convert = conv_2d(convert,32,5,activation='relu')
convert=max_pool_2d(convert,5)
convert = conv_2d(convert,64,5,activation='relu')
convert=max_pool_2d(convert,5)
convert = conv_2d(convert,128,5,activation='relu')
convert=max_pool_2d(convert,5)
convert = conv_2d(convert,64,5,activation='relu')
convert=max_pool_2d(convert,5)
convert = conv_2d(convert,32,5,activation='relu')
convert=max_pool_2d(convert,5)

convert = fully_connected(convert,1024,activation='relu')
convert=dropout(convert,0.8)

convert = fully_connected(convert,2,activation='softmax')
convert = regression(convert,optimizer='adam',learning_rate=LR,loss='categorical_crossentropy',name=)

model=tflearn.DNN(convert,tensorboard_dir='log',tensorboard_verbose=0)
model.fit({'input':X_train},{'targets':y_train},n_epoch=10,
         validation_set=({'input': X_test},{'targets':y_test}),
          snapshot_step=500,show_metric=True,run_id=MODEL_NAME)

fig=plt.figure(figsize=(16,12))

for num,data in enumerate(test_data[:16]):
    img_num = data[1]
    img_data = data[0]
    
    y = fig.add_subplot(4,4,num+1)
    orig = img_data
    data= img_data.reshape(IMG_SIZE,IMG_SIZE,1)
    model_out = model.predict([data])[0]
    
    if np.argmax(model_out) ==1:
        str_label = 'Dog'
        
    else:
        str_label ='Cat'
    y.imshow(orig,cmap='gray')
    y.axes.get_x_axis().set_visible(False)
    y_axes.get_yaxis().set_visible(False)
plt.show()
```
